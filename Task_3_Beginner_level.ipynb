# IMAGE - 1 : THE MARVEL AVENGERS

### Reading the image

```python
image1 = cv2.imread("avenger.jpg")
cv2.imshow("Avengers", image1)  #To display image on our windows pc
cv2.waitKey(0)
```

# IMAGE 2 -  PENCIL 

```python
image2 = cv2.imread("pencil.jpg")
cv2.imshow("Pencil", image2)
cv2.waitKey(0)
```

# IMAGE 3 - CHRISTMAS TREE

```python
image3 = cv2.imread("christmastree.jpg")
cv2.imshow("Christmas_tree", image3)
cv2.waitKey(0)
```

### Creating a new image by converting the original image to greyscale

```python
gray_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
cv2.imshow("Avengers_pencil_sketched", gray_image1)
cv2.waitKey(0)
```

```python
gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
cv2.imshow("Pencil_pencil_sketched", gray_image2)
cv2.waitKey(0)
```

```python
gray_image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2GRAY)
cv2.imshow("Christmas_tree_sketched", gray_image3)
cv2.waitKey(0)
```

### Inverting the new grayscale image

```python
inverted_image1 = 255 - gray_image1
cv2.imshow("Inverted1", inverted_image1)
cv2.waitKey()
```

```python
inverted_image2 = 255 - gray_image2
cv2.imshow("Inverted2", inverted_image2)
cv2.waitKey()
```

```python
inverted_image3 = 255 - gray_image3
cv2.imshow("Inverted2", inverted_image3)
cv2.waitKey()
```

### Bluring the image by using the Gaussian Function in OpenCV

```python
blurred1 = cv2.GaussianBlur(inverted_image1, (21, 21), 0)
```

```python
blurred2 = cv2.GaussianBlur(inverted_image2, (21, 21), 0)
```

```python
blurred3 = cv2.GaussianBlur(inverted_image3, (21, 21), 0)
```

### Inverting the blurred image and then we can easily convert the image into a pencil sketch

```python
inverted_blurred1 = 255 - blurred1
pencil_sketch1 = cv2.divide(gray_image1, inverted_blurred1, scale=256.0)
cv2.imshow("Pencil Sketch1", pencil_sketch1)
cv2.waitKey(0)
```

```python
inverted_blurred2 = 255 - blurred2
pencil_sketch2 = cv2.divide(gray_image2, inverted_blurred2, scale=256.0)
cv2.imshow("Pencil Sketch2", pencil_sketch2)
cv2.waitKey(0)
```

```python
inverted_blurred3 = 255 - blurred3
pencil_sketch3 = cv2.divide(gray_image3, inverted_blurred3, scale=256.0)
cv2.imshow("Pencil Sketch3", pencil_sketch3)
cv2.waitKey(0)
```

### looking at both the original image and the pencil sketch

```python
cv2.imshow("original image1", image1)
cv2.imshow("pencil sketch1", pencil_sketch1)
cv2.waitKey(0)
```

```python
cv2.imshow("original image2", image2)
cv2.imshow("pencil sketch2", pencil_sketch2)
cv2.waitKey(0)
```

```python
cv2.imshow("original image3", image3)
cv2.imshow("pencil sketch3", pencil_sketch3)
cv2.waitKey(0)
```
